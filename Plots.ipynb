```python
# Import our modules that we are using
import matplotlib.pyplot as plt
import numpy as np
from pylab import figure, text, scatter, show
import numpy as np
import matplotlib.pyplot as plt
```

```python
def main():
    a = -1
    b = 1

    y, x = np.ogrid[-5:5:100j, -5:5:100j]
    plt.contour(x.ravel(), y.ravel(), pow(y, 2) - pow(x, 3) - x * a - b, [0], colors = 'black')
    plt.grid()
    plt.text(-3.9, 3, '$y^2$ = $x^3$ - x + 1', size=20, color='black')
    fig1 = plt.gcf()
    plt.show()
    plt.draw()
    fig1.savefig('./Plots/ec_plot1.png')

if __name__ == '__main__':
    main()
```

```python
def main():
    a = -1
    b = 3

    y, x = np.ogrid[-5:5:100j, -5:5:100j]
    plt.contour(x.ravel(), y.ravel(), pow(y, 2) - pow(x, 3) - x * a - b, [0], colors = 'black')
    plt.grid()
    #plt.text(-3.9, 3, '$y^2$ = $x^3$ - x + 3', size=20, color='black')
    plt.text(-2, 0.7, 'P', size=20, color='black')
    plt.text(-0.4, 2.1, 'Q', size=20, color='black')
    plt.text(1.5, 3.2, 'R', size=20, color='black')
    plt.text(1.2, -3.4, '-R', size=20, color='black')
    fig1 = plt.gcf()
    plt.plot([-4, 4], [-1, 4.4], 'k-', lw=2)
    plt.plot([2, 2], [-4, 4], 'k--', lw=2)
    plt.text(-3.9, 4.2, 'P + Q = -R', size=20, color='black')
    plt.plot(label = "Real Discriminator Loss")
    plt.show()
    plt.draw()
    fig1.savefig('./Plots/ec_plot2.png')

if __name__ == '__main__':
    main()
```

```python
def main():
    a = -1
    b = 3

    y, x = np.ogrid[-5:5:100j, -5:5:100j]
    plt.contour(x.ravel(), y.ravel(), pow(y, 2) - pow(x, 3) - x * a - b, [0], colors = 'black')
    plt.grid()
    #plt.text(-3.9, 3, '$y^2$ = $x^3$ - x + 3', size=20, color='black')
    plt.text(-1.2, 2, 'P', size=20, color='black')
    #plt.text(-0.4, 2.1, 'Q', size=20, color='black')
    plt.text(1.3, -1.9, '-G', size=20, color='black')
    plt.text(1.3, 1.4, 'G', size=20, color='black')
    plt.text(2.4, -3.8, '-R', size=20, color='black')
    fig1 = plt.gcf()
    plt.plot([-2.1, 3.1], [3.5, -5], 'k-', lw=2)
    plt.plot([1.2, 1.2], [-4, 4], 'k--', lw=2)
    plt.text(-3.9, 4.2, 'P + P + P = G', size=20, color='black')
    plt.show()
    plt.draw()
    fig1.savefig('./Plots/ec_plot4.png')

if __name__ == '__main__':
    main()
```

```python
# Loss Values

with open('Losses/noisy4layers_50epochs_5000nodes_256batch_size_0.001lr_0.0var_weight/average_td_gd_loss.txt') as f:
    d_loss = f.read()   
    
d_loss = d_loss.strip('][').split(', ')
d_loss = [float(v) for v in d_loss]


with open('Losses/noisy4layers_50epochs_5000nodes_256batch_size_0.001lr_0.0var_weight/g_d_running_loss.txt') as f:
    f_d_loss = f.read()   
    
f_d_loss = f_d_loss.strip('][').split(', ')
f_d_loss = [float(v) for v in f_d_loss]


with open('Losses/noisy4layers_50epochs_5000nodes_256batch_size_0.001lr_0.0var_weight/running_loss_true_d.txt') as f:
    r_d_loss = f.read()   
    
r_d_loss = r_d_loss.strip('][').split(', ')
r_d_loss = [float(v) for v in r_d_loss]


with open('Losses/noisy4layers_50epochs_5000nodes_256batch_size_0.001lr_0.0var_weight/running_loss.txt') as f:
    g_loss = f.read()   
    
g_loss = g_loss.strip('][').split(', ')
g_loss = [float(v) for v in g_loss]


plt.plot(g_loss, label = "Generator Loss")
plt.plot(d_loss, label = "Discriminator Loss")
plt.plot(f_d_loss, label = "Fake Discriminator Loss")
plt.plot(r_d_loss, label = "Real Discriminator Loss")
#plt.set_ylim([0, 15])
#plt.axis((0,50,0, 75))
plt.legend()
plt.savefig('./Plots/Final_Model_Losses.png')

```
